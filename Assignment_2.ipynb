```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Activation Func with Linear Units

```python
# plot inputs and outputs
x = np.arange(-10, 10, 0.1)
y = x
plt.plot(x,y)
plt.title('Linear Activation Function')
plt.xlabel('Input')
plt.ylabel('Output');
```

### 2. Activation Func with Sigmoid Units

```python
# plot inputs and outputs
import numpy as np 
def sig(x):
 return 1/(1 + np.exp(-x))

x = np.arange(-10, 10, 0.1)
y = sig(x)
plt.plot(x,y)
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output');
```

### 3. Activation Func with Hyperbolic Tangent Units

```python
x = np.arange(-10, 10, 0.1)
y = np.tanh(x)
plt.plot(x,y)
plt.title('Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output');
```

```python
x = np.arange(-10, 10, 0.1)
y = (2 / (1 + np.exp(-2*x)))-1
plt.plot(x,y)
plt.title('Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output');
```

### 4. Activation Func with Rectified Linear Units (ReLU) 

```python
x = np.arange(-10, 10, 0.1)
z = np.zeros(len(x))
y = np.maximum(z,x)
plt.plot(x,y)
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output');
```
